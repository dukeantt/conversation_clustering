```python
import plotly.graph_objects as go
import plotly.express as px
import pandas as pd


df = pd.read_csv("data/tsne_vectors_2d_input_text_only.csv")

# df['description'] = [""] * len(df)
# for index, value in df.iterrows():
#     if value['cluster'] == 2:
#         df.at[index,'description'] = "Greet"
#     elif value['cluster'] == 0:
#         df.at[index,'description'] = "Handle url failed"
#     elif value['cluster'] == 4:
#         df.at[index,'description'] = "Thank/Handover/Agree/Disagree/Address"
#     else:
#         df.at[index,'description'] = "Ask price, availability"

cluster_methods = ["kmeans_cluster","gm_cluster","hc_cluster"]
menus_list = {}
for cluster_method in cluster_methods:
    buttons_list = [dict(label="None",
                         method="relayout",
                         args=["shapes", []])]
    menus = []
    clusters = []
    for i in range(8):
        cluster_df = df[df[cluster_method] == i]
        x = cluster_df['x']
        y = cluster_df['y']
        x_min = x.min()
        y_min = y.min()
        x_max = x.max()
        y_max = y.max()
        cluster = [dict(type="circle",
                                xref="x", yref="y",
                                x0=x_min, y0=y_min,
                                x1=x_max, y1=y_max,
                                line=dict(color=px.colors.sequential.Plasma[i]))]
        clusters.append(cluster)
        button = dict(label="Cluster " + str(i),
                      method="relayout",
                      args=["shapes", cluster])
        buttons_list.append(button)

    all_cluster = []
    for item in clusters:
        all_cluster.append(item[0])

    buttons_list.append(dict(label="All",
                            method="relayout",
                            args=["shapes", all_cluster]))
    buttons_list = list(buttons_list)
    menus = dict(buttons=buttons_list)
    menus_list[cluster_method] = menus


fig = px.scatter(
    df,
    x='x',
    y='y',
    color="kmeans_cluster",
    hover_data=["short_input_texts", "short_bot_texts","intent","short_entities","action_1"],
    )

fig.update_layout(
    updatemenus=[menus_list['kmeans_cluster']]
)


fig.show()
```
